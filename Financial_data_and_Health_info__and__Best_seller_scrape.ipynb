Get quotes from cnbc

```
from bs4 import BeautifulSoup
import requests
import datetime

import pandas as pd

time_stamp = datetime.datetime.now()
html = requests.get('https://www.cnbc.com/quotes/?symbol=AAPL').content

soup = BeautifulSoup(html, 'html.parser')
symbol = soup.find(itemprop='tickerSymbol').get('content')
price = soup.find(itemprop="price").get('content')
price_change =  soup.find(itemprop='priceChange').get('content')
priceChangePercent = soup.find(itemprop='priceChangePercent').get('content')
#volume = soup.find('span', {'class' = '.ng-binding'}).get('content')
quote_time = time_stamp
df = pd.DataFrame({
     'symbol' : symbol,
     'price' : price,
     'price_change': price_change,
     'priceChangePercent' :priceChangePercent,
     'quote_time': quote_time}, index =[0])

print(df)
```

Repeat the code with timer

```
from bs4 import BeautifulSoup
import requests
import time

for i in range (10):
  html = requests.get(f'https://finance.yahoo.com/quote/AAPL' ).content
  soup = BeautifulSoup(html, 'html.parser')
  price = soup.find_all('div', {'class':'My(6px) Pos(r) smartphone_Mt(6px)'})[0].find('span').text
  print({'AAPL': price})
  time.sleep(10)
```
